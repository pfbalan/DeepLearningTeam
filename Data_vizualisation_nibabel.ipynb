```python
archive = tarfile.open('fs_t1_oasis.tar', 'r')
OASIS = archive.getnames()
archive2 = tarfile.open('fs_t1_nacc.tar', 'r')
NACC = archive2.getnames()
```

```python
count = 0
for i in OASIS:
    archive.extract(i, path="OASIS/")
    count+=1
    if count == 100:
        break
```
